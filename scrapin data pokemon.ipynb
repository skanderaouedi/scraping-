```python
from selenium import webdriver
import pandas as pd
```

```python
driver =webdriver.Chrome(executable_path='D:\chromedriver.exe')
driver.get('https://pokemondb.net/pokedex/all')
driver.maximize_window()
privesy=driver.find_element_by_xpath('//p[@class="text-right"]/button')
privesy.click()
```

```python
#get names 
names=driver.find_elements_by_xpath('//td[@class="cell-name"]/a[@class="ent-name"]')
#get type 
type=driver.find_elements_by_xpath('//td[@class="cell-icon"]')
#get total
total=driver.find_elements_by_xpath('//td[@class="cell-total"]')
#get hp
hp=driver.find_elements_by_xpath('//tr/td[@class="cell-num"][1]')
#get attack 
attack=driver.find_elements_by_xpath('//tr/td[@class="cell-num"][2]')
#get defense 
defense=driver.find_elements_by_xpath('//tr/td[@class="cell-num"][3]')
#get sp_atk
sp_atk=driver.find_elements_by_xpath('//tr/td[@class="cell-num"][4]')
#get sp_def
sp_def=driver.find_elements_by_xpath('//tr/td[@class="cell-num"][5]')
#get speed
speed=driver.find_elements_by_xpath('//tr/td[@class="cell-num"][6]')
```

```python
for x in speed:
    print(x.text )
```

```python
df=pd.DataFrame()
n=[]
t=[]
tot=[]
h=[]
at=[]
dfe=[]
st=[]
sd=[]
sde=[]

for x in names:
    n.append(x.text)
for x in type:
    t.append(x.text)
for x in total:
    tot.append(x.text)
for x in hp:
    h.append(x.text)
for x in attack:
    at.append(x.text)
for x in defense:
    dfe.append(x.text)
for x in sp_atk:
    st.append(x.text)
for x in sp_def:
    sd.append(x.text)
for x in speed:
    sde.append(x.text)
df['names']=n
df['type']=t
df['total']=tot
df['attack']=at
df['hp']=h
df['defense']=dfe
df['speed_attack']=st
df['speed_defense']=sd
df['speed']=sde
```

```python
#replace '\n' with a comma
df['type']=df['type'].apply(lambda x: x.replace("\n",','))
```
